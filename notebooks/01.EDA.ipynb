```python
# =============================
# 1. Import Libraries
# =============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Pandas settings
pd.set_option('display.max_columns', None)

# Matplotlib style handling for old/new versions
try:
    plt.style.use('seaborn-v0_8')  # Matplotlib ≥ 3.6
except:
    plt.style.use('seaborn')       # Older Matplotlib versions

# =============================
# 2. Load Dataset
# =============================
df = pd.read_csv(r"E:\Codes\Projects\ML\air_quality\data\city_day.csv")

# Preview dataset
print("Shape:", df.shape)
display(df.head())

# =============================
# 3. Basic Info
# =============================
display(df.info())
display(df.describe())

# =============================
# 4. Missing Values
# =============================
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing, 'Percentage': missing_pct})

display(missing_df)

# Plot missing values
plt.figure(figsize=(12,6))
missing_df['Percentage'].plot(kind='bar', color='skyblue')
plt.title('Missing Data (%)')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()

# =============================
# 5. Unique Cities
# =============================
print("Number of Cities:", df['City'].nunique())
print("Cities List:", df['City'].unique())

# Count plot of city data availability
plt.figure(figsize=(12,6))
df['City'].value_counts().plot(kind='bar', color='orange')
plt.title('Data Points per City')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# =============================
# 6. Date Column Analysis
# =============================
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print("Date Range:", df['Date'].min(), "to", df['Date'].max())

# =============================
# 7. AQI Distribution
# =============================
if df['AQI'].notnull().sum() > 0:
    plt.figure(figsize=(10,5))
    sns.histplot(df['AQI'].dropna(), bins=50, kde=True, color='green')
    plt.title('AQI Distribution')
    plt.xlabel('AQI')
    plt.tight_layout()
    plt.show()

# =============================
# 8. AQI Bucket Analysis
# =============================
if df['AQI_Bucket'].notnull().sum() > 0:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x='AQI_Bucket',
                  order=df['AQI_Bucket'].value_counts().index,
                  palette='viridis')
    plt.title('AQI Buckets Count')
    plt.tight_layout()
    plt.show()

# =============================
# 9. Time-Series Example for One City
# =============================
city_sample = "Delhi"  # Change to explore other cities
city_df = df[df['City'] == city_sample].sort_values('Date')

if not city_df.empty:
    plt.figure(figsize=(15,5))
    plt.plot(city_df['Date'], city_df['AQI'], marker='o', markersize=2, color='blue')
    plt.title(f"AQI Trend in {city_sample}")
    plt.xlabel("Date")
    plt.ylabel("AQI")
    plt.tight_layout()
    plt.show()

# =============================
# 10. Correlation Analysis (Pollutants)
# =============================
pollutants = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
corr = df[pollutants].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Pollutants and AQI")
plt.tight_layout()
plt.show()
```
